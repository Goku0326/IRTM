### Exploratory Data Anlysis

```python
def process_text(text):
    pattern = r'CHAPTER (\d+)\n\n([\s\S]*?)(?=CHAPTER \d+|$)'
    matches = re.findall(pattern, text)
    
    pattern1 = r'CHAPTER (\d+)\n\n(.*?)\n'
    matches1 = re.findall(pattern1, file1)
    chapter_numbers = []
    chapter_name = []
    for match in matches1:
        chapter_number1 = match[0]
        chapter_numbers.append(chapter_number1)
        chapter_name1 = match[1]
        chapter_name.append(chapter_name1)
     
    def extract_books(text):
        book_pattern = r"Book \d+: ([^\n]+)\n"
        book = re.findall(book_pattern, text)
        return book

    def extract_book_names(text):
        book_name_pattern = r'Book \d+'
        book_names = re.findall(book_name_pattern, text)
        return book_names

    book_names = extract_book_names(file1)
    books = extract_books(file1) 
    
    chapter_texts = []

    for match in matches:
        chapter_number = match[0]
        chapter_text = match[1]

        # Split the chapter text into lines
        chapter_lines = chapter_text.split('\n')

        chapter_texts.append(chapter_lines)

    data = []
    # Access and process each chapter text
    for chapter_number, chapter_text, chapter_name in zip(chapter_numbers, chapter_texts, chapter_name):
        # Tokenize the chapter text
        tokens = word_tokenize(' '.join(chapter_text))
        word_count = len(tokens)
        
        # Print the chapter text
        chapter_lines = '\n'.join(chapter_text)
        data.append([book_names, books, chapter_number, chapter_name, tokens[5:], word_count])
            
    # Create a dataframe
    df = pd.DataFrame(data, columns=['Book Number', 'Book Name', 'Chapter Number', 'Chapter Name', 'WordCloudText', 'WordCount'])
    return df
```
